```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision
import torchvision.transforms as transforms
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from collections import Counter

# Configuration
BATCH_SIZE = 128
LEARNING_RATE = 0.001
EPOCHS = 4

# Check for GPU
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")
```

```python
import torchvision.transforms as transforms

# Define the pipeline: Gray -> MinMax(0-1) -> Gaussian Blur -> Pad
transform_pipeline = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),                       # Handles MinMax scaling (0-1)

    # Gaussian Blur
    # kernel_size=(5, 5): The size of the blur window (Must be odd numbers: 3, 5, 7...)
    # sigma=(0.1, 2.0): The range of how "strong" the blur is.
    transforms.GaussianBlur(kernel_size=(5, 5), sigma=(0.1, 2.0)),

    transforms.Pad(padding=2, fill=0),           # Pad 28x28 -> 32x32
])

# Download and Load Data
train_dataset = torchvision.datasets.MNIST(root='./data', train=True,
                                           download=True, transform=transform_pipeline)
test_dataset = torchvision.datasets.MNIST(root='./data', train=False,
                                          download=True, transform=transform_pipeline)

train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print("Data Loaded Successfully.")
```

```python
# Iterating a single batch to inspect
data_iter = iter(train_loader)
images, labels = next(data_iter)

print("--- Tensor Inspection ---")
print(f"Batch Shape: {images.shape} (Batch, Channel, Height, Width)")
print(f"Label Shape: {labels.shape}")
print(f"Data Type:   {images.dtype}")
print(f"Min Value:   {images.min()} (Should be 0.0)")
print(f"Max Value:   {images.max()} (Should be <= 1.0)")
print("-" * 30)
```

```python
# 1. Visualize Class Frequency
targets = train_dataset.targets.numpy()
counter = Counter(targets)
plt.figure(figsize=(10, 4))
plt.bar(counter.keys(), counter.values(), color='skyblue')
plt.title('Frequency of Class Data')
plt.xlabel('Digit')
plt.xticks(range(10))
plt.show()

# 2. Visualize Samples
fig, axes = plt.subplots(1, 6, figsize=(12, 3))
for i in range(6):
    img = images[i].squeeze().numpy()
    axes[i].imshow(img, cmap='gray')
    axes[i].set_title(f"Label: {labels[i].item()}")
    axes[i].axis('off')
plt.suptitle("Pre-processed Samples (32x32)")
plt.show()
```

```python
# Input size: 32 * 32 = 1024
input_size = 32 * 32

model = nn.Sequential(
    nn.Flatten(),                 # Input Layer

    # Hidden Layer 1
    nn.Linear(input_size, 256),
    nn.ReLU(),

    # Hidden Layer 2
    nn.Linear(256, 128),
    nn.ReLU(),

    # Hidden Layer 3
    nn.Linear(128, 64),
    nn.ReLU(),

    # Output Layer
    nn.Linear(64, 10)
)

model = model.to(device)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)
train_losses = []

print("--- Starting Training ---")

for epoch in range(EPOCHS):
    model.train()
    running_loss = 0.0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        # Zero gradients
        optimizer.zero_grad()

        # Forward Pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward Pass
        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    epoch_loss = running_loss / len(train_loader)
    train_losses.append(epoch_loss)
    print(f"Epoch [{epoch+1}/{EPOCHS}] | Loss: {epoch_loss:.4f}")

print("Training Complete.")
```

```python
plt.figure(figsize=(8, 5))
plt.plot(train_losses, label='Training Loss', color='purple', linewidth=2)
plt.title('Training Loss Curve')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
model.eval()
all_preds = []
all_labels = []

# Disable gradient calculation for testing
with torch.no_grad():
    for images, labels in test_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        _, predicted = torch.max(outputs, 1)

        all_preds.extend(predicted.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

# Generate Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)

# Plot Heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens',
            xticklabels=range(10), yticklabels=range(10))
plt.title('Confusion Matrix (Test Data)')
plt.ylabel('True Label')
plt.xlabel('Predicted Label')
plt.show()
```

```python
def calculate_accuracy(model, loader, device):
    model.eval()  # Set model to evaluation mode (important!)

    correct_predictions = 0
    total_samples = 0

    # We do not need gradients for evaluation, so we turn them off to save memory/speed
    with torch.no_grad():
        for images, labels in loader:
            images, labels = images.to(device), labels.to(device)

            # Forward pass: Get the raw scores (logits)
            outputs = model(images)

            # Get the index of the highest score (the predicted class)
            # torch.max returns (values, indices) -> we only care about indices
            _, predicted = torch.max(outputs, 1)

            # Count how many predictions matched the true labels
            total_samples += labels.size(0)
            correct_predictions += (predicted == labels).sum().item()

    accuracy = 100 * correct_predictions / total_samples
    print(f"\n--- Accuracy Results ---")
    print(f"Correct Predictions: {correct_predictions} / {total_samples}")
    print(f"Final Accuracy:      {accuracy:.2f}%")
    return accuracy

# Run the function
calculate_accuracy(model, test_loader, device)
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
import torch.nn.functional as F
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.metrics import confusion_matrix
from sklearn.manifold import TSNE
import random

# Set seeds for reproducibility
torch.manual_seed(42)
np.random.seed(42)
random.seed(42)

# Configuration
BATCH_SIZE = 64
LEARNING_RATE = 0.001
EPOCHS = 10  # CNNs take a bit longer to train well
DEVICE = torch.device('cuda' if torch.cuda.is_available() else 'cpu')

print(f"Using Device: {DEVICE}")
```

```python
# --- 1. Training Transform (Augmentation) ---
train_transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    # Randomly rotate between -15 and +15 degrees
    transforms.RandomRotation(degrees=15),
    # Slight shifts (translation) by up to 10% of image height/width
    transforms.RandomAffine(degrees=0, translate=(0.1, 0.1)),
    transforms.ToTensor(),
    # Standardize MNIST mean/std (industry standard values)
    transforms.Normalize((0.1307,), (0.3081,))
])

# --- 2. Testing Transform (No Augmentation) ---
test_transform = transforms.Compose([
    transforms.Grayscale(num_output_channels=1),
    transforms.ToTensor(),
    transforms.Normalize((0.1307,), (0.3081,))
])

# Load Datasets with respective transforms
train_dataset = torchvision.datasets.MNIST(root='./data', train=True,
                                           download=True, transform=train_transform)

test_dataset = torchvision.datasets.MNIST(root='./data', train=False,
                                          download=True, transform=test_transform)

# Data Loaders
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=BATCH_SIZE, shuffle=True)

test_loader = torch.utils.data.DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)

print("Data Loaded. Training data will be randomly augmented on-the-fly.")
```

```python
def visualize_augmented_batch(loader):
    data_iter = iter(loader)
    images, labels = next(data_iter)

    plt.figure(figsize=(12, 4))
    for i in range(8):
        ax = plt.subplot(1, 8, i+1)
        # Un-normalize for visualization so it looks decent
        img = images[i] * 0.3081 + 0.1307
        plt.imshow(img.squeeze(), cmap='gray')
        plt.title(f"Label: {labels[i].item()}")
        plt.axis('off')
    plt.suptitle("Examples of On-the-Fly Augmentation (Rotation/Shifting)")
    plt.show()

visualize_augmented_batch(train_loader)
```

```python
class MNIST_CNN(nn.Module):
    def __init__(self):
        super(MNIST_CNN, self).__init__()

        # --- Block 1 ---
        # Input: [Batch, 1, 28, 28]
        # Conv2d(in_channels, out_channels, kernel_size, padding)
        # Padding=1 ensures output size remains same before pooling (28x28)
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, padding=1)
        self.bn1 = nn.BatchNorm2d(32)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2) # -> [32, 14, 14]

        # --- Block 2 ---
        self.conv2 = nn.Conv2d(32, 64, kernel_size=3, padding=1)
        self.bn2 = nn.BatchNorm2d(64)
        self.pool2 = nn.MaxPool2d(kernel_size=2, stride=2) # -> [64, 7, 7]

        # --- Classifier (Fully Connected) ---
        # Calculate input size for linear layer: 64 channels * 7 * 7 spatial size
        self.fc_input_size = 64 * 7 * 7
        self.flatten = nn.Flatten()
        self.fc1 = nn.Linear(self.fc_input_size, 128)
        self.dropout = nn.Dropout(0.5) # Prevents overfitting
        self.fc2 = nn.Linear(128, 10)  # 10 Output classes

    def forward(self, x):
        # Block 1 flow
        x = self.conv1(x)
        x = self.bn1(x)
        x = F.relu(x)
        x = self.pool1(x)

        # Block 2 flow
        x = self.conv2(x)
        x = self.bn2(x)
        x = F.relu(x)
        x = self.pool2(x)

        # Classifier flow
        x = self.flatten(x)
        x = self.fc1(x)
        x = F.relu(x)
        x = self.dropout(x)
        x = self.fc2(x)
        return x

model = MNIST_CNN().to(DEVICE)
print(model)
```

```python
criterion = nn.CrossEntropyLoss()
# Adam is great, but SGD with momentum often generalizes better for CNNs eventually.
# Let's stick to Adam for speed for now, it will still get 99%+
optimizer = optim.Adam(model.parameters(), lr=LEARNING_RATE)

train_losses = []
train_accs = []
test_losses = []
test_accs = []

print("--- Starting CNN Training ---")

for epoch in range(EPOCHS):
    # --- Training Phase ---
    model.train()
    running_loss = 0.0
    correct_train = 0
    total_train = 0

    for images, labels in train_loader:
        images, labels = images.to(DEVICE), labels.to(DEVICE)

        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total_train += labels.size(0)
        correct_train += (predicted == labels).sum().item()

    epoch_train_loss = running_loss / len(train_loader)
    epoch_train_acc = 100 * correct_train / total_train
    train_losses.append(epoch_train_loss)
    train_accs.append(epoch_train_acc)

    # --- Testing Phase (Validation) ---
    model.eval()
    running_test_loss = 0.0
    correct_test = 0
    total_test = 0

    with torch.no_grad():
        for images, labels in test_loader:
            images, labels = images.to(DEVICE), labels.to(DEVICE)
            outputs = model(images)
            loss = criterion(outputs, labels)

            running_test_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total_test += labels.size(0)
            correct_test += (predicted == labels).sum().item()

    epoch_test_loss = running_test_loss / len(test_loader)
    epoch_test_acc = 100 * correct_test / total_test
    test_losses.append(epoch_test_loss)
    test_accs.append(epoch_test_acc)

    print(f"Epoch [{epoch+1}/{EPOCHS}] | "
          f"Train Loss: {epoch_train_loss:.4f}, Acc: {epoch_train_acc:.2f}% | "
          f"Test Loss: {epoch_test_loss:.4f}, Acc: {epoch_test_acc:.2f}%")

print("Training Complete.")
```

```python
activation = {}
def get_activation(name):
    def hook(model, input, output):
        activation[name] = output.detach().cpu()
    return hook

# Register hook to the first pooling layer
handle = model.pool1.register_forward_hook(get_activation('pool1_output'))

# Run one image
data_iter = iter(test_loader)
images, labels = next(data_iter)
model.eval()
with torch.no_grad():
    _ = model(images[0].unsqueeze(0).to(DEVICE))

# Remove hook
handle.remove()

# Visualize the 32 feature maps extracted by Block 1
feature_maps = activation['pool1_output'].squeeze() # Shape: [32, 14, 14]

fig, axes = plt.subplots(4, 8, figsize=(16, 8))
fig.suptitle(f"Feature Maps after Block 1 (Input digit: {labels[0].item()})")
for i in range(32):
    ax = axes[i // 8, i % 8]
    ax.imshow(feature_maps[i], cmap='magma')
    ax.axis('off')
plt.show()
```

```python
# We need a partial model to get the output of fc1 (the 128 features)
class FeatureExtractor(nn.Module):
    def __init__(self, original_model):
        super().__init__()
        self.features = nn.Sequential(
            original_model.conv1, original_model.bn1, nn.ReLU(), original_model.pool1,
            original_model.conv2, original_model.bn2, nn.ReLU(), original_model.pool2,
            original_model.flatten,
            original_model.fc1, nn.ReLU() # Stop here, before dropout and final fc2
        )
    def forward(self, x):
        return self.features(x)

feature_model = FeatureExtractor(model).to(DEVICE)
feature_model.eval()

latent_features = []
latent_labels = []
print("Extracting features for t-SNE (this may take a minute)...")

with torch.no_grad():
    for i, (images, labels) in enumerate(test_loader):
        images = images.to(DEVICE)
        output = feature_model(images)
        latent_features.append(output.cpu().numpy())
        latent_labels.append(labels.numpy())
        if i * BATCH_SIZE >= 2000: break # Only use first 2000 samples to save time

X = np.concatenate(latent_features, axis=0)
y = np.concatenate(latent_labels, axis=0)

print(f"Running t-SNE on {X.shape} features...")
tsne = TSNE(n_components=2, random_state=42, perplexity=30)
X_embedded = tsne.fit_transform(X)

plt.figure(figsize=(10, 10))
scatter = plt.scatter(X_embedded[:, 0], X_embedded[:, 1], c=y, cmap='tab10', s=15, alpha=0.7)
plt.legend(*scatter.legend_elements(), title="Digits")
plt.title('t-SNE of CNN Latent Space (128 dense features -> 2D)')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
import torch

def calculate_accuracy(model, loader):
    # Fix 1: Automatically find out if the model is on CPU or GPU
    try:
        device = next(model.parameters()).device
    except Exception as e:
        print("Error: Could not detect model device. Is the model initialized?")
        return 0.0

    model.eval()  # Set to evaluation mode
    correct = 0
    total = 0

    print("Calculating accuracy... (this might take a moment)")

    with torch.no_grad(): # Fix 2: Disable gradients to prevent memory errors
        for images, labels in loader:
            # Fix 3: Move images/labels to the SAME device as the model
            images = images.to(device)
            labels = labels.to(device)

            # Forward pass
            outputs = model(images)

            # Get predictions
            # outputs.data shape: [Batch_Size, 10]
            # torch.max returns (max_value, index_of_max_value)
            _, predicted = torch.max(outputs.data, 1)

            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    if total == 0:
        print("Error: The data loader is empty!")
        return 0.0

    accuracy = 100 * correct / total
    print(f'Final Accuracy: {accuracy:.2f} %')
    return accuracy

# --- Run the function ---
# Make sure 'model' and 'test_loader' are defined before running this!
calculate_accuracy(model, test_loader)
```

```python
import matplotlib.pyplot as plt

# We use the same 'X_embedded' (3D data) and 'y' (labels) from before.

# 1. Setup the figure with 3 subplots
fig, axes = plt.subplots(1, 3, figsize=(20, 6))

# Define the planes and which columns of data to use
# Data columns: 0=x, 1=y, 2=z
projections = [
    (0, 1, 'XY Plane (Bottom View)'),  # x vs y
    (0, 2, 'XZ Plane (Side View 1)'),  # x vs z
    (1, 2, 'YZ Plane (Side View 2)')   # y vs z
]

# 2. Loop through and create each plot
for i, (dim1, dim2, title) in enumerate(projections):
    ax = axes[i]

    # Scatter plot for this projection
    scatter = ax.scatter(
        X_embedded[:, dim1],
        X_embedded[:, dim2],
        c=y,
        cmap='tab10',
        s=10,
        alpha=0.6
    )

    ax.set_title(title)
    ax.set_xlabel(f'Latent Dim {dim1+1}')
    ax.set_ylabel(f'Latent Dim {dim2+1}')
    ax.grid(True, alpha=0.3)

# 3. Add a single Legend for all plots
# We use the handles/labels from the last plot to make one common legend
handles, labels = scatter.legend_elements()
fig.legend(handles, labels, title="Digits", loc='center right')

plt.suptitle("Projections of 3D Latent Space onto 2D Planes", fontsize=16)
plt.subplots_adjust(right=0.9) # Make room for the legend
plt.show()
```
